```python
# FIXING OR CLEANING THE DATASET
# Identify all missing values in the dataset.
# Replace missing values in the discount column with 0.
# Calculate and fill missing values in total_sales using:
# total_sales = quantity × price − discount
# Remove duplicate rows from the dataset.
```

```python
df.isnull().sum()
```

```python
df['discount'] = df['discount'].fillna(0)
df['total_sales'] = df['total_sales'].fillna(df['quantity'] * df['price'] - df['discount'])
```

```python
df
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
# PREPROCESSING THE DATASET
# Create a new column discount_percent.
```

```python
df['discount_percent'] = (df['discount'] / df['price']) * 100
```

```python
# FEATURE ENGINEERING 
# Create a new column profit using:
#   profit = total_sales − (quantity × price × 0.8)
# Identify high-profit transactions (profit > 50000).
```

```python
df['profit'] = df['total_sales'] - (df['quantity'] * df['price'] * 0.8)
```

```python
# INSIGHTS
# Identify the month with highest sales.
# Identify the best-performing product.
# Identify which region generated the most revenue.
# Comment on how discount affects sales.
```

```python
# Identify the month with highest sales.
```

```python
month_sales = df.groupby('month')['total_sales'].sum()
month_sales
```

```python
highest_month = month_sales.idxmax()    # month with highest sales
highest_value = month_sales.max()       # value of highest sale

print("Month with highest sales: ",highest_month)
print("Value of highest sales: ",highest_value)
```

```python
# Identify the best-performing product (purchased more)
```

```python
products = df.groupby('product')['quantity'].sum()
products
```

```python
best_product = products.idxmax()
best_value = products.max()

print("Best Product: ",best_product)
print("Best Product value: ",best_value)
```

```python
# Identify which region generated the most revenue.
```

```python
region_revenue = df.groupby('region')['total_sales'].sum()
region_revenue
```

```python
highest_region = region_revenue.idxmax()
highest_value = region_revenue.max()

print("Region with max revenue: ",highest_region)
print("value of max revenue: ",highest_value)
```

```python
# Comment on how discount affects sales. 
```

```python
# Correlation b/w discount & sales:
#   if corr > 0 : Higher the discount, higher is the sales
#   if corr < 0 : Higher the discount, lower is the sales

correlation = df["discount"].corr(df["total_sales"])
print("Correlation between Discount and Sales:", correlation)
```
